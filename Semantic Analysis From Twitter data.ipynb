```python
import nltk
import json_lines
import pandas as pd
import numpy as np
from nltk import word_tokenize
from nltk.corpus import stopwords
from textblob import TextBlob
```

```python
tweets=[]
tweet=[]
with open("nepal-quake-2015-tweets.jsonl",'r')as f:
    for item in json_lines.reader(f):
        tweets.append(item['text'])
        tweet.append(pd.DataFrame(item).user.location)
```

```python
print((tweet[449]))
```

```python
no_of_tweet=len(tweets)
```

```python

cleaned_tweet=[]
```

```python
import re
for i in range(no_of_tweet): 
    tu = re.sub(r"http\S+", "", tweets[i])
    new_sent = re.sub(r'[^\w\s]', '', tu)
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(new_sent)
    new_sentence =[]
    for w in  word_tokens:
        if w not in stop_words:
           new_sentence.append(w)
    new_sentence = ' '.join(new_sentence)
    cleaned_tweet.append(new_sentence.lower()) #Processed text to string list
```

```python
sntmnt=[]
for i in range(no_of_tweet):
    blob=TextBlob(cleaned_tweet[i]).sentiment
    sntmnt.append(blob)
#blob=TextBlob('I hate you because I love you').sentiment
```
